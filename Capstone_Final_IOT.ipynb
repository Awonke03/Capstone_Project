## **Model Preparation**

```python
# Separate features and target variable
X = df.drop('label', axis=1)
y = df['label']

# Encode the target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply ADASYN to the whole dataset
adasyn = ADASYN()
X_resampled, y_resampled = adasyn.fit_resample(X_scaled, y_encoded)

# Display the label distribution after sampling
print("Label distribution after ADASYN sampling:\n", pd.Series(y_resampled).value_counts())

# Convert the resampled dataset to a DataFrame for easier indexing
X_resampled_df = pd.DataFrame(X_resampled)
y_resampled_df = pd.Series(y_resampled)

# Create three bootstrap test sets for A/B testing without replacement
bootstrap_samples = []
n_samples = 15000  # Each test set will have about 15000 records
remaining_indices = X_resampled_df.index.tolist()

for i in range(3):
    # Ensure we don't exceed the number of remaining indices
    if len(remaining_indices) < n_samples:
        print("Not enough data points left to create the test set.")
        break

    test_indices = np.random.choice(remaining_indices, size=n_samples, replace=False)
    X_bootstrap = X_resampled_df.loc[test_indices]
    y_bootstrap = y_resampled_df.loc[test_indices]
    bootstrap_samples.append((X_bootstrap, y_bootstrap))
    print(f"Test set {i+1} label distribution:\n", y_bootstrap.value_counts())

    # Remove the selected indices from the remaining indices to avoid reuse
    remaining_indices = list(set(remaining_indices) - set(test_indices))

# Use the remaining data as the training set
X_train_data = X_resampled_df.loc[remaining_indices]
y_train_data = y_resampled_df.loc[remaining_indices]

# Split the remaining resampled data into the final training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_train_data, y_train_data, train_size=0.8, shuffle=True, random_state=42)

# Display the label distribution in the training and test sets
print("Training label distribution:\n", y_train.value_counts())
print("Test label distribution:\n", y_test.value_counts())
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
# Initialize and train the Naive Bayes model
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Make predictions
nb_train_pred = nb_model.predict(X_train)
nb_test_pred = nb_model.predict(X_test)

# Calculate accuracies
nb_train_accuracy = accuracy_score(y_train, nb_train_pred)
nb_test_accuracy = accuracy_score(y_test, nb_test_pred)

# Bias-Variance Analysis (using accuracy-based method)

# Bias Calculation: Training error (1 - accuracy)
train_bias = 1 - nb_train_accuracy

# Variance Calculation: Difference between training and testing accuracy
variance = abs(nb_train_accuracy - nb_test_accuracy)

print("Naive Bayes Model")
print(f"Train Accuracy: {nb_train_accuracy:.4f}")
print(f"Test Accuracy: {nb_test_accuracy:.4f}")
print(f"Bias (Training Error): {train_bias:.4f}")
print(f"Variance (Generalization Gap): {variance:.4f}")
```

```python
brf_model = BalancedRandomForestClassifier(
    sampling_strategy='all',
    replacement=True,
    bootstrap=False
)
brf_model.fit(X_train, y_train)

# Make predictions
brf_train_pred = brf_model.predict(X_train)
brf_test_pred = brf_model.predict(X_test)

# Calculate accuracies
brf_train_accuracy = accuracy_score(y_train, brf_train_pred)
brf_test_accuracy = accuracy_score(y_test, brf_test_pred)

# Bias Calculation: Training error (1 - accuracy)
train_bias = 1 - brf_train_accuracy

# Variance Calculation: Difference between training and testing accuracy
variance = abs(brf_train_accuracy - brf_test_accuracy)

print("Balanced Random Forest Model")
print(f"Train Accuracy: {brf_train_accuracy:.4f}")
print(f"Test Accuracy: {brf_test_accuracy:.4f}")
print(f"Bias (Training Error): {train_bias:.4f}")
print(f"Variance (Generalization Gap): {variance:.4f}")
```

```python
# Initialize metrics dictionary
metrics_dict = {
    'Model': [],
    'Accuracy': [],
    'F1 Score': [],
    'Recall': [],
    'Precision': [],
    'AUC-ROC': []
}

# Function to compute metrics
def compute_metrics(y_true, y_pred, model_name):
    metrics_dict['Model'].append(model_name)
    metrics_dict['Accuracy'].append(accuracy_score(y_true, y_pred))
    metrics_dict['F1 Score'].append(f1_score(y_true, y_pred))
    metrics_dict['Recall'].append(recall_score(y_true, y_pred))
    metrics_dict['Precision'].append(precision_score(y_true, y_pred))
    metrics_dict['AUC-ROC'].append(roc_auc_score(y_true, y_pred))

# Compute metrics for each model
compute_metrics(y_test, nb_test_pred, 'Naive Bayes')
compute_metrics(y_test, brf_test_pred, 'Balanced Random Forest')
compute_metrics(y_test, xgb_test_pred, 'Extreme Gradient Boosting')
compute_metrics(y_test, ada_test_pred, 'AdaBoost')
compute_metrics(y_test, dnn_y_pred, 'Deep Neural Network')
compute_metrics(y_test, fnn_y_pred, 'Feedforward Neural Network')
compute_metrics(y_test, mlp_y_pred, 'Multilayer Perceptron')
compute_metrics(y_test, rnn_y_pred, 'Recurrent Neural Network')
```
